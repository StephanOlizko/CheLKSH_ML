```python
import pandas as pd

#Ваш код ниже...
```

```python
df = pd.read_csv('Titanic-Dataset-Processed.csv')

df["Cabin"].value_counts()
```

```python
#Удалим все числа из столбца Cabin
df['Cabin'] = df['Cabin'].str.replace('\d+', '', regex=True)

df["Cabin"].value_counts()
```

```python
a = set([j for i in df["Cabin"].unique() for j in i.split()])

a
```

```python
#Создадим новые столбцы для каждого значения из множества a
for i in a:
    df[i] = df["Cabin"].str.contains(i)

df
```

```python
df = df.drop(columns=["Cabin"])
```

```python
#Сохраним результат в файл
df.to_csv('Titanic-Dataset-Processed-1.csv' , index=False)
```
