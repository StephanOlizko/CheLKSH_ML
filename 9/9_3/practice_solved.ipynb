1. Постройте график функции y = x^2 и график функции y = 4x - 6 на одном поле. Определите абсциссы пересечений графиков, используя функцию plt.show()

```python
x = list(range(-10, 11))
y1 = [i**2 for i in x]
y2 = [4*i + 32 for i in x]
#...

#Добавим сетку на график
plt.grid()
#Обозначения на x через 1
plt.xticks(range(-10, 11, 1))

#...

plt.plot(x, y1)
plt.plot(x, y2)
plt.show()
```

2. Постройте график функции y = sin(x) в диапазоне от -2π до 2π

```python
#Постройте график функции y = sin(x) в диапазоне от -2π до 2π без numpy
import math
x = [i/10 for i in range(-63, 64)]
y = [math.sin(i) for i in x]
plt.plot(x, y)

plt.grid()
plt.axis('equal')
plt.show()


```

3. Нарисуйте окружность с центром в точке (0, 0) и радиусом 5 красным пунктиром
___

```python
x = [i/10 for i in range(-50, 51)]
y = [(25 - i**2)**0.5 for i in x]
plt.plot(x, y, "r--")
plt.plot(x, [-i for i in y], "r--")
#...


plt.axis('equal')
plt.show()
```
