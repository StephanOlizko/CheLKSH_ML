## Учимся работать с matplotlib

Это огромный модуль для визуализации графиков. Охватывает всевозможную математику и статистику. Большинство других модулей визуализации работают на его основе. 

Чтобы построить график, нужно вызвать функцию plot. В нее передается то, что мы хотим визуализировать. Это может быть набор значений, функция, которая задает график и дополнительные параметры. Также существует ряд функций, который упрощает чтение графика, например, можно подписать оси и дать графику название.

```python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4])
plt.show()
```

Если в функцию plot передается просто набор чисел, то matplotlib считает, что ему передали значения переменной y. Что хранится в переменной x, он додумывает сам. По умолчанию подбирается набор значений такой же длины, что и переданный, только отсчет идет от 0, как у всего в Python. Именно поэтому на получившемся графике максимальный х = 3. Сделаем так, чтобы все было корректно.

```python
plt.plot([1, 2, 3, 4], [1, 2, 3, 4])
plt.scatter(2, 3)
plt.show()
```

matplotlib построит любой график просто по набору точек.

```python
# "r--" означает красный цвет и пунктирную линию
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], "r--")
plt.show()
```

Аналогичным образом можно задать не просто какую-то кривую, а график функции. Например, построим $y(x) = x^2$.

```python
x = list(range(0, 20)) # задаем значения х
y = [x1**2 for x1 in x] # считаем у(х)
plt.figure(figsize=(10, 10))
plt.plot(x, y, label="y(x)=x^2") # строим график
plt.xlabel("x") # подписываем ось х
plt.ylabel("y") # подписываем ось у
plt.title("График квадратичной функции") # подписываем график
plt.legend()# добавляем легенду
plt.grid(True)
plt.show()
```



На одном графике можно изобрать несколько функций. Например, y(x)=x и y(x)=x2.


```python
x = list(range(0, 10)) # задаем значения х
y1 = [x1**2 for x1 in x] # считаем значения для квадратичной функции
y2 = [x1 for x1 in x]
_, ax = plt.subplots(figsize=(5, 10)) # создаем график для нескольких кривых
ax.plot(x, y2, label="y(x)=x") # строим график линейной функции
ax.plot(x, y1, label="y(x)=x^2") # строим график квадратичной функции
ax.set_xlabel("x") # подписываем ось х
ax.set_ylabel("y") # подписываем ось у
ax.set_title("Графики функций") # подписываем график
ax.grid(True) # добавляем сетку
plt.legend() # добавляем легенду
plt.show()
```

И можно на одной картинке уместить несколько графиков.


```python
x = list(range(0, 10)) # задаем значения х
y1 = [x1**2 for x1 in x] # считаем значения для квадратичной функции
y2 = [x1**3 for x1 in x] # считаем значения для кубической функции

_, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(10, 3)) # создаем график для нескольких кривых
ax1.plot(x, x, "tab:pink") # строим график линейной функции
ax1.set_title("Линейная функция")
ax2.plot(x, y1, "tab:green") # строим график квадратичной функции
ax2.set_title("Квадратичная функция")
ax3.plot(x, y2, label="y(x)=x^3") # строим график кубической функции
ax3.set_title("Кубическая функция")
plt.show()
```

### Делаем лабораторку по физике с matplotlib

Лабораторная работа: Исследование зависимости периода колебаний маятника от его длины

Цель работы:

Изучить зависимость периода колебаний математического маятника от его длины и подтвердить теоретическую зависимость T=2π * sqrt(L/g)​

Оборудование:

    Стенд с маятником
    Линейка или измерительная лента
    Секундомер
    Набор грузов
    Калькулятор

Ход работы:

    Закрепите маятник на штативе.
    Измерьте длину нити маятника L с помощью линейки или измерительной ленты.
    Отклоните маятник на небольшой угол (около 5-10 градусов) и отпустите его.
    С помощью секундомера измерьте время t, за которое маятник совершает 10 полных колебаний.
    Рассчитайте период одного колебания T по формуле T=t/10 
    Повторите измерения для разных длин L (например, 30 см, 40 см, 50 см, 60 см, 70 см).
    Запишите результаты в таблицу.

```python
#Таблица измерений:
#Длина нити (L), м	Время (t) для 10 колебаний, с	Период (T), с
#0.30	         11.0	                         1.10
#0.40	         12.6	                         1.26
#0.50	         14.0	                         1.40
#0.60	         15.4	                         1.54
#0.70	         16.6	                         1.66

d = {"L": [0.3, 0.4, 0.5, 0.6, 0.7],
    "t": [11.0, 12.6, 14.0, 15.4, 16.6],
    "T": [1.1, 1.26, 1.4, 1.54, 1.66]}

df = pd.DataFrame(data=d)

df
```

```python
lengths = df["L"]
periods = df["T"]
times = df["t"]


# Построение графика
plt.figure(figsize=(8, 6))
plt.plot(lengths, periods, marker='o', linestyle='-', color='b')
plt.title('Зависимость периода колебаний маятника от длины')
plt.xlabel('Длина нити (м)')
plt.ylabel('Период колебаний (с)')
plt.grid(True)
plt.show()
```

Расчет погрешностей

Для измерений в физике важно учитывать погрешности. Рассчитаем погрешности измерений периода колебаний и длины нити.

Погрешности:

Погрешность измерения длины нити ΔL: примем ΔL=±0.01 м (точность линейки).

Погрешность измерения времени Δt: примем Δt=±0.2 с (реакция человека при использовании секундомера).

Погрешность периода колебаний:

Погрешность периода ΔT можно рассчитать по формуле:
ΔT=Δ/t10

```python
periods = times / 10
length_errors = 0.01  # Погрешность измерения длины
time_errors = 0.2     # Погрешность измерения времени
period_errors = time_errors / 10  # Погрешность периода

# Рассчитываем квадраты периодов
periods_squared = periods ** 2
periods_squared_errors = 2 * periods * period_errors  # Погрешность квадратов периодов
```

```python
import numpy as np

def linear_fit_with_errors(x, y, x_errors, y_errors):
    # Средние значения
    x_mean = np.mean(x)
    y_mean = np.mean(y)
    
    # Средние значения с учетом погрешностей
    x_mean_plus_error = np.mean(x + x_errors)
    y_mean_plus_error = np.mean(y + y_errors)
    x_mean_minus_error = np.mean(x - x_errors)
    y_mean_minus_error = np.mean(y - y_errors)
    
    # Находим коэффициенты для основной прямой
    coefficients_main = np.polyfit(x, y, 1)
    
    # Находим коэффициенты для прямой с верхней погрешностью
    coefficients_upper = np.polyfit(x + x_errors, y + y_errors, 1)
    
    # Находим коэффициенты для прямой с нижней погрешностью
    coefficients_lower = np.polyfit(x - x_errors, y - y_errors, 1)
    
    return coefficients_main, coefficients_upper, coefficients_lower

# Находим уравнения прямых для основной, верхней и нижней линий
coeffs_main, coeffs_upper, coeffs_lower = linear_fit_with_errors(lengths, periods, length_errors, period_errors)

# Создаем линейные функции для построения прямых
linear_main = np.poly1d(coeffs_main)
linear_upper = np.poly1d(coeffs_upper)
linear_lower = np.poly1d(coeffs_lower)

# Параметры для графиков
x_fit = np.linspace(min(lengths), max(lengths), 100)
y_main = linear_main(x_fit)
y_upper = linear_upper(x_fit)
y_lower = linear_lower(x_fit)
```

```python
# График зависимости периода колебаний от длины нити с "покачанными" прямыми
plt.figure(figsize=(8, 6))
plt.errorbar(lengths, periods, yerr=period_errors, xerr=length_errors, fmt='o', ecolor='r', capsize=5, label='Экспериментальные данные')
plt.plot(x_fit, y_main, 'b-', label='Основная прямая')
plt.plot(x_fit, y_upper, 'g--', label='Верхняя прямая')
plt.plot(x_fit, y_lower, 'm--', label='Нижняя прямая')
plt.title('Зависимость периода колебаний маятника от длины')
plt.xlabel('Длина нити (м)')
plt.ylabel('Период колебаний (с)')
plt.grid(True)
plt.legend()
plt.show()

# График зависимости квадрата периода колебаний от длины нити с "покачанными" прямыми
coeffs_main_squared, coeffs_upper_squared, coeffs_lower_squared = linear_fit_with_errors(lengths, periods_squared, length_errors, periods_squared_errors)

linear_main_squared = np.poly1d(coeffs_main_squared)
linear_upper_squared = np.poly1d(coeffs_upper_squared)
linear_lower_squared = np.poly1d(coeffs_lower_squared)

y_main_squared = linear_main_squared(x_fit)
y_upper_squared = linear_upper_squared(x_fit)
y_lower_squared = linear_lower_squared(x_fit)

plt.figure(figsize=(8, 6))
plt.errorbar(lengths, periods_squared, yerr=periods_squared_errors, xerr=length_errors, fmt='o', ecolor='r', capsize=5, label='Экспериментальные данные')
plt.plot(x_fit, y_main_squared, 'b-', label='Основная прямая')
plt.plot(x_fit, y_upper_squared, 'g--', label='Верхняя прямая')
plt.plot(x_fit, y_lower_squared, 'm--', label='Нижняя прямая')
plt.title('Зависимость квадрата периода колебаний маятника от длины')
plt.xlabel('Длина нити (м)')
plt.ylabel('Квадрат периода колебаний (с^2)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
def calculate_errors(coeffs_upper, coeffs_lower):
    delta_a = abs(coeffs_upper[0] - coeffs_lower[0]) / 2
    delta_b = abs(coeffs_upper[1] - coeffs_lower[1]) / 2
    return delta_a, delta_b

# Погрешности для периода
delta_a_main, delta_b_main = calculate_errors(coeffs_upper, coeffs_lower)

# Погрешности для квадрата периода
delta_a_squared, delta_b_squared = calculate_errors(coeffs_upper_squared, coeffs_lower_squared)

print(f"Коэффициенты основной прямой для периода: a = {coeffs_main[0]:.4f}, b = {coeffs_main[1]:.4f}")
print(f"Погрешности коэффициентов: Δa = {delta_a_main:.4f}, Δb = {delta_b_main:.4f}")

print(f"Коэффициенты основной прямой для квадрата периода: a = {coeffs_main_squared[0]:.4f}, b = {coeffs_main_squared[1]:.4f}")
print(f"Погрешности коэффициентов: Δa = {delta_a_squared:.4f}, Δb = {delta_b_squared:.4f}")
```
